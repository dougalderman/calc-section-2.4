```r
x<-c(0,1,2,3,4,5,6)
y<-c(10,8,9.5,15,12,14,14.5)
```

```r
plot(x,y,col="blue", main="Pressure of a Pregnant Woman's Contractions as a Function of Time",xlab="Time (Minutes)", ylab="Pressure (mm of Hg)")
```

```r
relation<-lm(y~x)
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=0,to=6, by=0.1)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
prediction <- predict(relation, data.frame(x=7))
```

```r
prediction
```
